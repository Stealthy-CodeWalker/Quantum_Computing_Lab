```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, array_to_latex, plot_histogram
import qiskit.quantum_info as qi
from qiskit_aer import Aer
```

```python
#PART A
qc = QuantumCircuit(1)
qc.x(0)
display(qc.draw('mpl'))
stv1 = qi.Statevector.from_instruction(qc)
print(stv1.data)
print(stv1.draw())
display(stv1.draw('latex',prefix='Statevector: '))
display(array_to_latex(stv1,prefix = "Matrix: "))
display(plot_bloch_multivector(stv1))
```

```python
#CNOT ON |00>
qc = QuantumCircuit(2)
qc.cx(0,1)
display(qc.draw('mpl'))
qc.measure_all()
backend = Aer.get_backend('aer_simulator')
result = backend.run(qc).result()
count = result.get_counts()
display(plot_histogram(count))
```
